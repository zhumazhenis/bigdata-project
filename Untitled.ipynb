#### 1) Following columns should be non-zero values, therefore zeros will be replaced by mean.
1. Glucose
2. BloodPressure
3. SkinThickness
4. Insulin
5. BMI

```python
def replaceWithMeans(data):
    means = {}
    nonZeroColumns = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
    for column in nonZeroColumns:
        count = 0
        s = 0
        for value in data[column]:
            if value != 0:
                count += 1
                s += value
        means[column] = s / count

    # replace with mean
    for k, v in means.items():
        data[k].replace(to_replace=[0], value=means[k], inplace=True)

replaceWithMeans(data)
data.head()
```

### Choose best scalers: Standard, MinMax, Robust

```python
# Standard scale
def standardScale(data, featureColumns):
#     featureColumns = ['Pregnancies', 'Glucose', 'BloodPressure', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Insulin', 'SkinThickness']
    scaledFeaturesData = StandardScaler().fit_transform(data[featureColumns])
    scaledData = pd.DataFrame(scaledFeaturesData, columns=featureColumns)
    scaledData['Outcome'] = data['Outcome']
    return scaledData

# MinMax scale
def minMaxScale(data, featureColumns):
#     featureColumns = ['Pregnancies', 'Glucose', 'BloodPressure', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Insulin', 'SkinThickness']
    scaledFeaturesData = MinMaxScaler().fit_transform(data[featureColumns])
    scaledData = pd.DataFrame(scaledFeaturesData, columns=featureColumns)
    scaledData['Outcome'] = data['Outcome']
    return scaledData

# Robust scale
def robustScale(data, featureColumns):
#     featureColumns = ['Pregnancies', 'Glucose', 'BloodPressure', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Insulin', 'SkinThickness']
    scaledFeaturesData = RobustScaler().fit_transform(data[featureColumns])
    scaledData = pd.DataFrame(scaledFeaturesData, columns=featureColumns)
    scaledData['Outcome'] = data['Outcome']
    return scaledData


```

### Scaler Comparison

```python
def compareScalers(modelsMap, data):
    result = {
        'Classifier': [],
        'Standard': [],
        'MinMax': [],
        'Robust': []
    }
    for name, model in modelsMap.items():
        scoring = make_scorer(f1_score)
        
        # StandardScaler
        featureColumns = ['Pregnancies', 'Glucose', 'BloodPressure', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Insulin', 'SkinThickness']
        outputColumn = 'Outcome'
        standardScaledData = standardScale(data, featureColumns)
        trainData, testData = train_test_split(standardScaledData, test_size=0.20, random_state=66, stratify=standardScaledData['Outcome'])
        
        trainFeatures, trainOutcome = trainData[featureColumns], trainData[outputColumn]
        testFeatures, testOutcome = testData[featureColumns], testData[outputColumn]
        crossValidationResult = cross_val_score(model, trainFeatures, trainOutcome, cv = 5, scoring = scoring)
        result['Standard'].append(crossValidationResult.mean())
        
        # MinMaxScaler
        featureColumns = ['Pregnancies', 'Glucose', 'BloodPressure', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Insulin', 'SkinThickness']
        outputColumn = 'Outcome'
        minMaxScaledData = minMaxScale(data, featureColumns)
        trainData, testData = train_test_split(minMaxScaledData, test_size=0.20, random_state=66, stratify=minMaxScaledData['Outcome'])
        
        trainFeatures, trainOutcome = trainData[featureColumns], trainData[outputColumn]
        testFeatures, testOutcome = testData[featureColumns], testData[outputColumn]
        crossValidationResult = cross_val_score(model, trainFeatures, trainOutcome, cv = 5, scoring = scoring)
        result['MinMax'].append(crossValidationResult.mean())
        
        # RobustScaler
        featureColumns = ['Pregnancies', 'Glucose', 'BloodPressure', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Insulin', 'SkinThickness']
        outputColumn = 'Outcome'
        robustScaledData = robustScale(data, featureColumns)
        trainData, testData = train_test_split(robustScaledData, test_size=0.20, random_state=66, stratify=robustScaledData['Outcome'])
        
        trainFeatures, trainOutcome = trainData[featureColumns], trainData[outputColumn]
        testFeatures, testOutcome = testData[featureColumns], testData[outputColumn]
        crossValidationResult = cross_val_score(model, trainFeatures, trainOutcome, cv = 5, scoring = scoring)
        result['Robust'].append(crossValidationResult.mean())
    
        result['Classifier'].append(name)        
    scalerComparison = pd.DataFrame(result)
    return scalerComparison

modelsMap = {
    'SVC': svm.SVC(), 
    'RandomForestClassifier': RandomForestClassifier(), # has feature_importances_
    'DecisionTreeClassifier': DecisionTreeClassifier(), # has feature_importances_
    'GaussianNB': GaussianNB(), 
    'KNeighborsClassifier': KNeighborsClassifier()
}

compareScalers(modelsMap, data)
```

### Hyper Parameter Tuning 

```python
# Split Data
featureColumns = ['Pregnancies', 'Glucose', 'BloodPressure', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Insulin', 'SkinThickness']
outputColumn = 'Outcome'
scaledData = standardScale(data, featureColumns)
trainData, testData = train_test_split(scaledData, test_size=0.20, random_state=66, stratify=scaledData['Outcome'])

xTrain, yTrain = trainData[featureColumns], trainData[outputColumn]
xTest, yTest = testData[featureColumns], testData[outputColumn]

hyperParametersMap = {
    'SVC': {
        'kernel': ['rbf', 'linear', 'poly', 'sigmoid'],
        'degree': [2, 3, 4],
        'tol': [1e-2, 1e-3, 1e-4]
    }, 
    'RandomForestClassifier': {
        'n_estimators': np.arange(1, 30, 2),
        'max_depth': np.arange(1, 10),
        'max_features': [1, 2, 3, 4, 5, 6, 7, 8, 'auto', 'log2', None],
        'bootstrap': [True, False],
    },
    'DecisionTreeClassifier': {
        'criterion': ['gini', 'entropy'],
        'max_depth': np.arange(1, 11),
        'max_features': [1, 2, 3, 4, 5, 6, 7, 8, 'auto', 'log2', None]
    },
    'GaussianNB': {
        'priors': [[0.05 * i, 0.05 * (20 - i)] for i in range(2, 19)]
    }, 
#     'GradientBoostingClassifier': {
#         'n_estimators': np.arange(10, 150, 10),
#         'max_depth': np.arange(1, 10, 2),
#         'max_features': [1, 2, 3, 4, 5, 6, 7, 8, 'auto', 'log2', None]
#     },
    'KNeighborsClassifier': {
        'n_neighbors': np.arange(3, 16),
        'p': [1, 2],
    }
}

modelsMap = {
    'SVC': svm.SVC(), 
    'RandomForestClassifier': RandomForestClassifier(), # has feature_importances_
    'DecisionTreeClassifier': DecisionTreeClassifier(), # has feature_importances_
    'GaussianNB': GaussianNB(), 
#     'GradientBoostingClassifier': GradientBoostingClassifier(), # has feature_importances_
    'KNeighborsClassifier': KNeighborsClassifier()
}

def findBestHyperParameters(xTrain, yTrain, hyperParametersMap, modelsMap):
    bestHyperParameters = {}
    for k, v in hyperParametersMap.items():
        bestHyperParameters[k] = {}
        clf = modelsMap[k]
        grid = GridSearchCV(estimator=clf, param_grid=v, cv=5, scoring=make_scorer(f1_score))
        grid.fit(xTrain, yTrain)
        bestHyperParameters[k]['f1Score'] = grid.best_score_
        bestHyperParameters[k]['parameters'] = grid.best_params_
    return bestHyperParameters


def bestHyperParametersToDataFrame(bestHyperParameters):
    bestHyperParametersDataFrame = pd.DataFrame(columns=['Classifier', 'F1 Score CV', 'Parameters'])
    for k, v in bestHyperParameters.items():
        vals = {
            'Classifier': k,
            'F1 Score CV': v['f1Score'],
            'Parameters': str(v['parameters'])
        }
        bestHyperParametersDataFrame = bestHyperParametersDataFrame.append(vals, ignore_index=True)
    return bestHyperParametersDataFrame.sort_values(by=['F1 Score CV'], ascending=False).reset_index(drop=True)

bestHyperParameters = findBestHyperParameters(xTrain, yTrain, hyperParametersMap, modelsMap)

pd.set_option('max_colwidth', -1)
bestHyperParametersToDataFrame(bestHyperParameters)
```

### Test classifiers with best hyperparameters

```python
modelsMap = {
    'SVC': svm.SVC, 
    'RandomForestClassifier': RandomForestClassifier, # has feature_importances_
    'DecisionTreeClassifier': DecisionTreeClassifier, # has feature_importances_
    'GaussianNB': GaussianNB, 
#     'GradientBoostingClassifier': GradientBoostingClassifier(), # has feature_importances_
    'KNeighborsClassifier': KNeighborsClassifier
}

for k, v in bestHyperParameters.items():
    clf = modelsMap[k](**v['parameters'])
    clf = clf.fit(xTrain, yTrain)
    yPred = clf.predict(xTest)
    print('Classifier: %s' % (k))
    print('F1 Score: %s' % (v['f1Score']))
    print(classification_report(yTest, yPred))
    print(confusion_matrix(yTest, yPred))
    #print(clf.feature_importances_)
    print('======================================================\n\n')
        
```

#### Age column will divied into three categorical columns Young, Middle, Old
According to https://www.healthline.com/health/type-2-diabetes-age-of-onset#risk-factors-foradults , people who are older than 45 are more likely to be diabetes.

```python
trainData, testData = train_test_split(data, test_size=0.20, random_state=66, stratify=data['Outcome'])
trainData = trainData.reset_index(drop=True)
testData = testData.reset_index(drop=True)

uniqueAges = trainData['Age'].unique()
uniqueAges.sort()

def findAllAgePivots(uniqueAges, data):
    n = len(uniqueAges)
    result = [[None] * n for i in range(n)]
    for i in range(n - 2):
        for j in range(i + 1, n - 1):
            age1 = uniqueAges[i]
            age2 = uniqueAges[j]
            zeros = data[data['Age'] <= age1]['Outcome'].value_counts().get(0)
            ones = data[data['Age'] <= age1]['Outcome'].value_counts().get(1)
            zeros = zeros if zeros != None else 0
            ones = ones if ones != None else 0
            young = (zeros, ones)
            
            zeros = data[(age1 < data['Age']) & (data['Age'] <= age2)]['Outcome'].value_counts().get(0)
            ones = data[(age1 < data['Age']) & (data['Age'] <= age2)]['Outcome'].value_counts().get(1)
            zeros = zeros if zeros != None else 0
            ones = ones if ones != None else 0
            middle = (zeros, ones)
            
            zeros = data[age2 < data['Age']]['Outcome'].value_counts().get(0)
            ones = data[age2 < data['Age']]['Outcome'].value_counts().get(1)
            zeros = zeros if zeros != None else 0
            ones = ones if ones != None else 0
            old = (zeros, ones)
            
            result[i][j] = [young, middle, old]
    return result


def findBestAgePivot(allAgePivots):
    n = len(allAgePivots)
    largest = 0
    ans = ()
    for i in range(n):
        for j in range(n):

            vals = allAgePivots[i][j]
            if vals == None:
                continue
            temp = 0

            youngCount = sum(vals[0])
            zeros, ones = vals[0][0], vals[0][1]
            temp = temp + 100 * abs(ones - zeros)

            middleCount = sum(vals[1])
            zeros, ones = vals[1][0], vals[1][1]
            temp = temp + 100 * abs(ones - zeros)

            oldCount = sum(vals[2])
            zeros, ones = vals[2][0], vals[2][1]
            temp = temp + 100 * abs(ones - zeros)

            temp = temp / (youngCount + middleCount + oldCount)
            if largest < temp:
                largest = temp
                ans = (i, j)

    return (largest, ans)

allAgePivots = findAllAgePivots(uniqueAges, trainData)
bestAgePivot = findBestAgePivot(allAgePivots)

left = bestAgePivot[1][0]
right = bestAgePivot[1][1]
print('Young: %d - %d ages, Class count: ' %(uniqueAges[0], uniqueAges[left]), allAgePivots[left][right][0])
print('Middle: %d - %d ages, Class count: ' %(uniqueAges[left] + 1, uniqueAges[right]), allAgePivots[left][right][1])
print('Old: %d - %d ages, Class count: ' %(uniqueAges[right] + 1, uniqueAges[-1]), allAgePivots[left][right][2])
```
