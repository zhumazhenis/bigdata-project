#### 1) Following columns should be non-zero values, therefore zeros will be replaced by mean.
1. Glucose
2. BloodPressure
3. SkinThickness
4. Insulin
5. BMI

```python
def replaceWithMeans(data):
    means = {}
    nonZeroColumns = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
    for column in nonZeroColumns:
        count = 0
        s = 0
        for value in data[column]:
            if value != 0:
                count += 1
                s += value
        means[column] = s / count

    # replace with mean
    for k, v in means.items():
        data[k].replace(to_replace=[0], value=means[k], inplace=True)

replaceWithMeans(data)
data.head()
```

### Split data

```python
def splitData(data):
    trainData, testData = train_test_split(data, test_size=0.20, random_state=66, stratify=data['Outcome'])
    featureColumns = ['Pregnancies', 'Glucose', 'BloodPressure', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Insulin', 'SkinThickness']
    outputColumn = 'Outcome'
    trainFeatures = trainData[featureColumns]
    testFeatures = testData[featureColumns]
    trainOutcome = trainData[outputColumn]
    testOutcome = testData[outputColumn]
    
    return trainFeatures, trainOutcome, testFeatures, testOutcome

xTrain, yTrain, xTest, yTest = splitData(data)
```

### Chost best scalers: Standard, MinMax, Robust

```python
# Standard scale
def standardScale(data):
    featureColumns = ['Pregnancies', 'Glucose', 'BloodPressure', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Insulin', 'SkinThickness']
    scaledFeaturesData = StandardScaler().fit_transform(data[featureColumns])
    scaledData = pd.DataFrame(scaledFeaturesData, columns=featureColumns)
    scaledData['Outcome'] = data['Outcome']
    return scaledData

# MinMax scale
def minMaxScale(data):
    featureColumns = ['Pregnancies', 'Glucose', 'BloodPressure', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Insulin', 'SkinThickness']
    scaledFeaturesData = MinMaxScaler().fit_transform(data[featureColumns])
    scaledData = pd.DataFrame(scaledFeaturesData, columns=featureColumns)
    scaledData['Outcome'] = data['Outcome']
    return scaledData

# Robust scale
def robustScale(data):
    featureColumns = ['Pregnancies', 'Glucose', 'BloodPressure', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Insulin', 'SkinThickness']
    scaledFeaturesData = RobustScaler().fit_transform(data[featureColumns])
    scaledData = pd.DataFrame(scaledFeaturesData, columns=featureColumns)
    scaledData['Outcome'] = data['Outcome']
    return scaledData
```

### Scaler Comparison

```python
def compareScalers(modelsMap, data):
    result = {
        'Classifier': [],
        'Standard': [],
        'MinMax': [],
        'Robust': []
    }
    for name, model in modelsMap.items():
        scoring = make_scorer(f1_score)
        # StandardScaler
        minMaxScaledData = minMaxScale(data)
        trainFeaturesMinMax, trainOutcomeMinMax, testFeaturesMinMax, testOutcomeMinMax = splitData(minMaxScaledData)
        crossValidationResult = cross_val_score(model, trainFeaturesMinMax, trainOutcomeMinMax, cv = 5, scoring = scoring)
        result['Standard'].append(crossValidationResult.mean())
        
        # MinMaxScaler
        minMaxScaledData = minMaxScale(data)
        trainFeaturesMinMax, trainOutcomeMinMax, testFeaturesMinMax, testOutcomeMinMax = splitData(minMaxScaledData)
        crossValidationResult = cross_val_score(model, trainFeaturesMinMax, trainOutcomeMinMax, cv = 5, scoring = scoring)
        result['MinMax'].append(crossValidationResult.mean())
        
        # RobustScaler
        robustScaledData = robustScale(data)
        trainFeaturesRobust, trainOutcomeRobust, testFeaturesRobust, testOutcomeRobust = splitData(robustScaledData)
        crossValidationResult = cross_val_score(model, trainFeaturesRobust, trainOutcomeRobust, cv = 5, scoring = scoring)
        result['Robust'].append(crossValidationResult.mean())
    
        result['Classifier'].append(name)        
    scalerComparison = pd.DataFrame(result)
    return scalerComparison

modelsMap = {
    'SVC': svm.SVC(), 
    'RandomForestClassifier': RandomForestClassifier(), # has feature_importances_
    'DecisionTreeClassifier': DecisionTreeClassifier(), # has feature_importances_
    'GaussianNB': GaussianNB(), 
    'GradientBoostingClassifier': GradientBoostingClassifier(), # has feature_importances_
    'KNeighborsClassifier': KNeighborsClassifier()
}

compareScalers(modelsMap, data)
```

Almost all scalers works the same, but let's choose Standard Scaler, which performs well for Euclidean Distance.

### Hyper Parameter Tuning 

```python
# Split Data
data = standardScale(data)
xTrain, yTrain, xTest, yTest = splitData(data)

hyperParametersMap = {
    'SVC': {
        'kernel': ['rbf', 'linear', 'poly', 'sigmoid'],
        'degree': [2, 3, 4],
        'tol': [1e-2, 1e-3, 1e-4]
    }, 
    'RandomForestClassifier': {
        'n_estimators': np.arange(1, 10, 2),
        'max_depth': np.arange(1, 5),
        'max_features': [1, 2, 3, 4, 5, 6, 7, 8, 'auto', 'log2', None],
        'bootstrap': [True, False],
    },
    'DecisionTreeClassifier': {
        'criterion': ['gini', 'entropy'],
        'max_depth': np.arange(1, 11),
        'max_features': [1, 2, 3, 4, 5, 6, 7, 8, 'auto', 'log2', None]
    },
    'GaussianNB': {
        'priors': [[0.05 * i, 0.05 * (20 - i)] for i in range(2, 19, 2)]
    }, 
#     'GradientBoostingClassifier': {
#         'n_estimators': np.arange(10, 150, 10),
#         'max_depth': np.arange(1, 10, 2),
#         'max_features': [1, 2, 3, 4, 5, 6, 7, 8, 'auto', 'log2', None]
#     },
    'KNeighborsClassifier': {
        'n_neighbors': np.arange(3, 16),
        'p': [1, 2],
    }
}

modelsMap = {
    'SVC': svm.SVC(), 
    'RandomForestClassifier': RandomForestClassifier(), # has feature_importances_
    'DecisionTreeClassifier': DecisionTreeClassifier(), # has feature_importances_
    'GaussianNB': GaussianNB(), 
    'GradientBoostingClassifier': GradientBoostingClassifier(), # has feature_importances_
    'KNeighborsClassifier': KNeighborsClassifier()
}

def findBestHyperParameters(xTrain, yTrain, hyperParametersMap, modelsMap):
    bestHyperParameters = {}
    for k, v in hyperParametersMap.items():
        bestHyperParameters[k] = {}
        clf = modelsMap[k]
        grid = GridSearchCV(estimator=clf, param_grid=v, cv=5, scoring=make_scorer(f1_score))
        grid.fit(xTrain, yTrain)
        bestHyperParameters[k]['f1Score'] = grid.best_score_
        bestHyperParameters[k]['parameters'] = grid.best_params_
    return bestHyperParameters


def bestHyperParametersToDataFrame(bestHyperParameters):
    bestHyperParametersDataFrame = pd.DataFrame(columns=['Classifier', 'F1 Score CV', 'Parameters'])
    for k, v in bestHyperParameters.items():
        vals = {
            'Classifier': k,
            'F1 Score CV': v['f1Score'],
            'Parameters': str(v['parameters'])
        }
        bestHyperParametersDataFrame = bestHyperParametersDataFrame.append(vals, ignore_index=True)
    return bestHyperParametersDataFrame.sort_values(by=['F1 Score CV'], ascending=False).reset_index(drop=True)

bestHyperParameters = findBestHyperParameters(xTrain, yTrain, hyperParametersMap, modelsMap)

pd.set_option('max_colwidth', -1)
bestHyperParametersToDataFrame(bestHyperParameters)
```
